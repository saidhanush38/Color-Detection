```python
# Import required libraries
import cv2
import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# Reading the image
img = cv2.imread("image path")
```

```python
# Initializing required variables
clicked = False
r = g = b = xpos = ypos = 0

# Importing the dataset
data = pd.read_csv('colors.csv')
```

```python
#Viewing the dataset
data.head()
```

```python
# Dividing the dataset as dependent and Independent variables
x = data.iloc[:,3:6].values
y = data[['hex','color']].values

# Splitting the data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)

# Initializing the classifier
knn = KNeighborsClassifier(n_neighbors=1)

# Training the model
knn.fit(x_train,y_train)

# Testing the model
y_pred = knn.predict(x_test)
```

```python
# Getting the x,y co-ordinates the mouse click

def get_xy(event, x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b,g,r,xpos,ypos, clicked
        clicked = True
        xpos = x
        ypos = y
        b,g,r = img[y,x]
        b = int(b)
        g = int(g)
        r = int(r)

cv2.namedWindow('image')
cv2.setMouseCallback('image',get_xy)


while(1):

    cv2.imshow("image",img)
    if (clicked):
        cv2.rectangle(img,(20,20), (1000,60), (b,g,r), -1)
        
        #cv2.rectangle(image, startpoint, endpoint, color, thickness)-1 fills entire rectangle 
        cv2.rectangle(img,(20,20), (1000,60), (b,g,r), -1)

        #Creating text string to display( Color name and RGB values )
        text = "The model predicts as " + knn.predict([[r,g,b]])[0,1] + ' R='+ str(r) +  ' G='+ str(g) +  ' B='+ str(b)
        
        #cv2.putText(img,text,start,font(0-7),fontScale,color,thickness,lineType )
        cv2.putText(img, text,(50,50),2,0.8,(255,255,255),2,cv2.LINE_AA)

        #For very light colours we will display text in black colour
        if(r+g+b>=600):
            cv2.putText(img, text,(50,50),2,0.8,(0,0,0),2,cv2.LINE_AA)
            
        clicked=False
        

    #Break the loop when user hits 'esc' key    
    if cv2.waitKey(20) & 0xFF ==27:
        break
    
cv2.destroyAllWindows()

```
